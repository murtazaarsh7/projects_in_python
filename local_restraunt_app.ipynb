```python
def add_to_cart(sku,quantity=1):
  if sku not in MENU:
    print('SORRY ITEM IS NOT EXIST,TRY AGAIN!')
    return
  if sku in MENU:
      if sku in CART:
        CART[sku]+=quantity
      else:
        CART[sku]=quantity

  print(f'ADDED ITEM:{MENU[sku]['NAME']},QUANTITY={quantity}')
```

```python

def modify_cart(sku, quantity):
  # First, check if the SKU itself exists in the MENU
  if sku not in MENU:
    print('SORRY, ITEM IS NOT EXIST IN MENU!')
    return

  if sku not in CART:
    # If the item is not in the cart, it can only be added with a positive quantity.
    if quantity > 0:
      CART[sku] = quantity
      print(f'Added {quantity} of {MENU[sku]['NAME']} to cart.')
    else:
      # Trying to remove or set to zero an item not in the cart.
      print(f'Cannot modify {MENU[sku]['NAME']} (not in cart) with quantity {quantity}.')
  else: # sku is in CART
    current_quantity = CART[sku]
    CART[sku] += quantity
    if CART[sku] <= 0:
      # If quantity becomes zero or negative, remove from cart
      CART.pop(sku)
      print(f'Removed {MENU[sku]['NAME']} from cart. Remaining quantity: 0.')
    else:
      print(f'Modified quantity of {MENU[sku]['NAME']} from {current_quantity} to {CART[sku]}.')
```

```python
def view_cart():
  print('\n***CART CONTENTS***\n')
  subtotal=0
  for sku in CART:
    subtotal+=MENU[sku]['PRICE']*CART[sku]
    print(f'item : {MENU[sku]['NAME']} x {CART[sku]} :: {MENU[sku]['PRICE']*CART[sku]}')
  total=subtotal+subtotal*SALES_TAX_RATE
  print(f'TOTAL CART VALUE::::{subtotal} + (10%)GST({subtotal*SALES_TAX_RATE})={total}')
```
